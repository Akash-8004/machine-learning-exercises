```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df = df.drop(["PassengerId", "Name", "SibSp", "Parch", "Parch", "Cabin", "Embarked", "Ticket"], axis =1)
```

```python
df.head()
```

```python
df[["male", "Female"]] = pd.get_dummies(df.Sex, dtype=int)
df = df.drop(["Sex","male"], axis=1)
```

```python
df.head(10)
```

```python
df["Age"].fillna(df["Age"].mean(), inplace=True)
```

```python
input = df.drop("Survived", axis=1)
target = df.Survived
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input, target, test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
np.average(cross_val_score(GaussianNB(),X_train, y_train, cv=10))
```
