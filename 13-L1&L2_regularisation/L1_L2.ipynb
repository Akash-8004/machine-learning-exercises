## Why use L1 & L2 Regularization?

1. Problem: In machine learning, sometimes the model learns too much from training data (overfitting). That means it becomes like a student who      memorizes the book but cannot answer new questions.

2. Solution: Regularization is like telling the model — “Don’t depend too much on any single feature. Keep it simple.”

- L1 Regularization (Lasso):

1. It makes some weights exactly zero.

2. This means it automatically removes unimportant features.

3. Easy to remember: L1 → 1 (zero out some weights → feature selection).

- L2 Regularization (Ridge):

1. It makes weights smaller but not zero.

2. This means the model uses all features, but none of them dominate.

3. Easy to remember: L2 → 2 (keep all, just shrink).

We are going to use Melbourne House Price Dataset where we'll predict House Predictions based on various features.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv("Melbourne_housing_FULL.csv")
dataset.head()
```

Let works with relevent columns

```python
dataset = dataset[['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']]
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
dataset[['Bedroom2', 'Bathroom', 'Car']] = dataset[['Bedroom2', 'Bathroom', 'Car']].fillna(0)
```

```python
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isna().sum()
```

```python
dataset.shape
```

one hot encode for categorical features

```python
dataset = pd.get_dummies(dataset, drop_first=True,dtype=int)
```

```python
dataset.head()
```

**data cleaning is done**
- lets split dataset into train an test

```python
X = dataset.drop('Price', axis=1)
y = dataset.Price
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3
)
```

### Now train dataset on linear regression model
- check accuracy with test dataset

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X, y)
```

```python
model.score(X_train, y_train)
```

```python
len(X_train)
```

**Normal Regression is clearly overfitting the data, let's try other models
Using Lasso (L1 Regularized) Regression Model**

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_test, y_test)
```

```python
lasso_reg.score(X_train, y_train)
```

### Using Ridge (L2 Regularized) Regression Model

```python
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(X_train, y_train)
```

```python
ridge_reg.score(X_test, y_test)
```

```python
ridge_reg.score(X_train, y_train)
```

## We see that Lasso and Ridge Regularizations prove to be beneficial when our Simple Linear Regression Model overfits. These results may not be that contrast but significant in most cases.Also that L1 & L2 Regularizations are used in Neural Networks too
